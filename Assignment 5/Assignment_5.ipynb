# Standardizing Numerical Columns

```python
numerical = dataset.select_dtypes(include=['int64', 'float64'])
numerical.head(7)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_num = pd.DataFrame(scaler.fit_transform(numerical), columns = numerical.columns)
df_num.head()
```

# Encoding Categorical Columns

```python
categorical = dataset.select_dtypes(include=['object'])
categorical.head(7)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse=False)
df_cat = pd.DataFrame(encoder.fit_transform(categorical),columns=encoder.get_feature_names(categorical.columns))
df_cat.head()
```

# Combining Numerical And Categorical columns

```python
adult = pd.concat([df_num, df_cat], axis=1)
adult.head(7)
```

```python
y = adult.pop('income_>50K')
x = adult
```

## Note :
- The target variable i.e., the income column is stored in y variable.
- All the data is stored in x variable.

## Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=100)
```

- Here the data is splited such that 70% of the data is taken for preparing the model, while 30% of the data is for testing purpose.

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Performing logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=500)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

temp_df.head(20)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

- It is seen that after applying logistic regression algorithm on the adult data to predict income of the person, 85% of accuracy is achieved.

```python
con_metric = metrics.confusion_matrix(y_test, y_pred)
con_metric
```

```python
from sklearn.metrics import precision_recall_fscore_support

precision_recall_fscore_support(y_test,y_pred,average='micro')
```

### This is the final classification report on the data

```python
print(metrics.classification_report(y_test,y_pred))
```

## - Using `RFE` to extract the best features and applying `Logistic Regression` on best features:-
* The `Recursive Feature Elimination` (RFE) method works by recursively removing attributes and building a model on those attributes that remain.
* It uses accuracy metric to `rank` the feature according to their importance.

```python
from sklearn.feature_selection import RFE

rfe = RFE(classifier,30)
rfe = rfe.fit(X_train, y_train)
```

```python
temp_df = pd.DataFrame({'Columns' : X_train.columns, 'Included' : rfe.support_, 'Ranking' : rfe.ranking_})

print(temp_df)
```

```python
imp_col = X_train.columns[rfe.support_]
imp_col
```

```python
X_train_new = X_train[imp_col]
X_train_new.head()
```

## Applying the best feature we will Train the model on Training Data

```python
from sklearn.linear_model import LogisticRegression
classifier_new = LogisticRegression(max_iter=250)
classifier_new.fit(X_train_new, y_train)
```

```python
X_test_new = X_test[imp_col]
X_test_new.head()
```

```python
y_pred_new = classifier_new.predict(X_test_new)
```

```python
temp_df_new = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_new})

temp_df_new.head(20)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_new)
```

# Observation :
- It is observed that by using only 30 features out of 96 features we have achieved 84% of accuracy.
- While using 96 feature we were able to achieve 85% of acuuracy.
- Therefore using RFE we have eliminated a lot of data and still we are abe to achieve accuracy very close to it.
- Hence RFE is a very good technique.

```python
print(metrics.classification_report(y_test,y_pred_new))
```

## - Using PCA to extract the best features :

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(x)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cummulative variance')
plt.show()
```

### Based on the graph it is seen that by using 30 features we can able to achieve 95% of accuracy

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca_new = PCA(n_components = 30)
pca_new_data = pca_new.fit_transform(x)
pca_new_data.shape
```

## -Train Test Split on the best features :

```python
x_train, x_test, y_train, y_test = train_test_split(pca_new_data[:], y, test_size=.30)
```

## A. `Logistic Regression` with PCA :

```python
logist = LogisticRegression(max_iter = 250)
logist.fit(x_train,y_train)

##Predicting the test data
y_predict_logistic = logist.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_predict_logistic)
```

```python
metrics.confusion_matrix(y_test,y_predict_logistic)
```

```python
print(metrics.classification_report(y_test,y_predict_logistic))
```

## B. `Support Vector Machine` with PCA :

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(x_train,y_train)
##Predicting the test data
y_predict_svm = svm.predict(x_test)
```

```python
print('Accuracy score : {}'.format(metrics.accuracy_score(y_test,y_predict_svm)))
print('\n')
print('Confusion matrix :')
print(metrics.confusion_matrix(y_test,y_predict_svm))
print('\n')
print('Classification Report :')
print(metrics.classification_report(y_test,y_predict_svm))
```

## C. `KNeighborsClassifier` with PCA :

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
##Predicting the test data
y_predict_knn = knn.predict(x_test)
```

```python
print('Accuracy score : {}'.format(metrics.accuracy_score(y_test,y_predict_knn)))
print('\n')
print('Confusion matrix :')
print(metrics.confusion_matrix(y_test,y_predict_knn))
print('\n')
print('Classification Report :')
print(metrics.classification_report(y_test,y_predict_knn))
```

## D. `Decision Tree Classifier` with PCA :

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
##Predicting the test data
y_predict_dtc = dtc.predict(x_test)
```

```python
print('Accuracy score : {}'.format(metrics.accuracy_score(y_test,y_predict_dtc)))
print('\n')
print('Confusion matrix :')
print(metrics.confusion_matrix(y_test,y_predict_dtc))
print('\n')
print('Classification Report :')
print(metrics.classification_report(y_test,y_predict_dtc))
```
